# Gravitational Effects on Particles

The following document seeks to analyze how particles behave in a uniform gravitational field.  The particles tested were a general particle of mass 1, a proton, and an electron.  The particles were given an initial velocity of $10 \, m/s$ upwards, then experienced uniform downward acceleration due to gravity.  The gravitational constant was chosen to be $-9.8 \, m/s$, as the gravitational acceleration in a vaccuum at sea level on earth.

```python
g = elementary.Particle(0, 0, 1)
p = elementary.Proton(0, 0, 2)
e = elementary.Electron(0, 0, 3)
# Each particle is given the same velocity
g.momentum = (0, 0, 10*g.mass)
p.momentum = (0, 0, 10*p.mass)
e.momentum = (0, 0, 10*e.mass)
```

The particles have two main properties: mass and momentum.  The desired velocity was transformed into an initial momentum by multiplying by the particle's mass, and each particle was given a slightly different starting point.  To increment the velocity, the formula $\Delta \vec{p} = \vec{a}*m*dt$, where $m$ is the mass of the particle and $dt$ is a constant, small change in time.  The program then evaluated all values between a time of 0 and a time of 5, then used them to plot the motion.

```python
# Initializing variables
dt = 0.01
t = 0
tPoints = []
gPoints = []
pPoints = []
ePoints = []
# Loops over each variable, delivering an impulse for each step of dt
while t < 5:
    g.impulse(0, 0, -9.8*g.mass*dt)
    g.move(dt)
    gPoints.append(g.position[2])
    p.impulse(0, 0, -9.8*p.mass*dt)
    p.move(dt)
    pPoints.append(p.position[2])
    e.impulse(0, 0, -9.8*e.mass*dt)
    e.move(dt)
    ePoints.append(e.position[2])
    tPoints.append(t)
    t += dt
f = pt.figure(figsize=(8,6))
a = pt.axes()
# Graphs the image, with three separate lines for each data set.  All data sets depend on t for the x-axis
a.plot(tPoints, gPoints, label = "General Particle")
a.plot(tPoints, pPoints, label = "Proton")
a.plot(tPoints, ePoints, label = "Electron")
a.set(xlabel = "Time (s)", ylabel = "Position Z-Axis (m)", title = "Particle Motion")
a.legend()
pt.show()
```

The graph very cleanly shows the correlation between the time elapsed and position on the z axis.  Most notably, the particles begin with some displacement from each other and relatively maintain that displacement.  The particles appear to become closer together as the time proceeds; however, this is merely due to the facct that the relative displacement of the particles is much smaller when the position is very large than when the position is small.  This verifies our expectations, which would be that motion due to gravity is not affected by the mass of the particles involved.
